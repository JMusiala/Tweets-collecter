```python
import snscrape.modules.twitter as snTwitter
import pandas as pd
import json
import datetime
```

```python
users = ["chelseafc","CFCDaily","FabrizioRomano","CFCBlues_com","CFCPys"]
counter = len(users)

tweets = []
limit = 50

def Likes(tweet):
    return tweet[2]

for i in range(counter):    

    sub_tweets = []
    scraper = snTwitter.TwitterUserScraper(users[i],False)
    n =0
    
    now = datetime.date.today()
    today = now.strftime("%d/%m/%Y")
    #print(today)
    for i,tweet in enumerate(scraper.get_items()):
        if n==10:
            break
        if len(tweets) == limit:
            break
        if today != tweet.date.strftime("%d/%m/%Y"):
            break
        else:
            tweets.append([tweet.date.strftime("%d/%m/%Y, %H:%M:%S"),tweet.user.username, tweet.likeCount , tweet.url ,tweet.content,
                          tweet.media])
            n=n+1
            tweets.sort(reverse=True,key = Likes)

df = pd.DataFrame(tweets, columns=['Date: ', 'User: ', 'Likes: ', 'Link: ', 'Tweet: ','Image: '])

df.to_html("tweety.html")
print(df)
```
